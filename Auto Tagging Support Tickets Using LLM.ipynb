```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification")
```

```python
labels = [
    "billing issue",
    "technical issue",
    "account problem",
    "refund request",
    "delivery issue",
    "women issue",
    "understanding issue"
]
```

```python
text = "My jazzcash payment was deducted but the order didn't confirm"
result = classifier(text, labels)
print(result['labels'][:3])
```

```python
text = "she is being stubborn"
result = classifier(text, labels)
print(result['labels'][:3])
```

```python
def zero_shot_predict(text):
  result = classifier(text, labels)
  return result['labels'][:3]
```

```python
tickets = [
    "My wifi is not working properly",
    "I wanna return my order",
    "I can't login to my insta account",
    "payment was deducted twice",
    "Delivery is delayed",
    "She is having difficulty in covering her syllabus"
]

for t in tickets:
  print("\nTickets: ",t)
  print("Tags: ", zero_shot_predict(t))
```

```python
def few_shot_predict(text):
    prompt = f"""
    Classify the support ticket into categories.

    Examples:
    "My internet is slow" → technical issue
    "I want refund" → refund request
    "Cannot login" → account problem

    Now classify:
    "{text}"
    """

    result = classifier(prompt, labels)
    return result['labels'][:3]
```

```python
for t in tickets:
    print("\nTicket:", t)
    print("Zero-shot:", zero_shot_predict(t))
    print("Few-shot:", few_shot_predict(t))
```

```python
result = classifier(text, labels)

for label, score in zip(result['labels'], result['scores']):
    print(label, ":", score)
```
